```python
import pandas as pd
import numpy as np 

from numpy.random import randn
```

```python
df = pd.DataFrame(randn(5,4), index = ["A","B","C","D","E"], columns = "W X Y Z".split())
df
```

```python
# Filtrando com iloc e trazendo somente informações de 1 coluna:
df.iloc[:5,:1] # O primeiro ":" é para linhas e o segunto é para colunas.
```

```python
# Filtrando com iloc e trazendo a primeira linha em todas as colunas: 

df.iloc[:1,:]
```

```python
# Filtrando com loc e trazendo toda as colunas:

df.loc[['A']]
```

```python
# Filtrando com loc e trazendo somento uma coluna:

df.loc[['A'],['W']]
```

```python
# Filtro por condições: 

df > 0
```

```python
# Filtro por condições retornando os valores:
df[df>0]
```

```python
df
```

```python
# Filtrando a coluna "W" se ela for maior que 0:

df[df['W']>0] # retorna os indices que forem verdadeiros.
```

```python
# Filtrando a coluna "W" se ela for maior que 0 e a coluna "Y" também se for maior que 0:

df[(df["W"] > 0) & (df["Y"]>0)]
```

```python
# Retornando o resultado boleano:

((df["W"]>0) & (df["Y"]>0))
```
